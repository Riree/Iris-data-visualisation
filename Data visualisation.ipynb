Comparison between various species based on sepal length and width using respective visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x="Species",data=df)
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm",hue="Species",data=df)
plt.title('Comparison of Species Based on Sepal Dimensions')
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()
```

Perform the Bi-variate Analysis via pairplot **visualization**

```python
sns.pairplot(df, hue='Species', vars=['SepalLengthCm', 'SepalWidthCm'])
plt.suptitle('Bivariate Analysis Using Pairplot', y=1.02)
plt.show()
```

```python
correlation = df[['SepalLengthCm', 'SepalWidthCm']].corr()
correlation
```

This weak negative correlation suggests that these features are likely independent of each other in terms of linear relationships (SepalLengthCm and SepalWidthCm as independent features.)

 4. **Visualize the categorical feature Species with all other four input variables**

```python
sns.pairplot(df,hues='Species',vars=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])
plt.suptitle('Pairplot of Species and Input Variables', y=1.02)
plt.show()
```

5. Plot the density of the length and width in the species.**bold text**

```python
fig, axes = plt.subplots(2, 2, figsize=(16,10))
sns.violinplot( y='SepalLengthCm', x= 'Species', data=df, orient='v' , ax=axes[0, 0],inner='quartile')
sns.violinplot( y='SepalWidthCm', x= 'Species', data=df, orient='v' , ax=axes[0, 1],inner='quartile')
sns.violinplot( y='PetalLengthCm', x= 'Species', data=df, orient='v' , ax=axes[1,0],inner='quartile')
sns.violinplot( y='PetalWidthCm', x= 'Species', data=df, orient='v' , ax=axes[1,1],inner='quartile')
plt.show()
```

6. Plot the probability density function(PDF) with each feature as a variable on X-axis and
its histogram and corresponding kernel density plot on Y-axis.

```python
sns.histplot(df['SepalLengthCm'], kde=True, bins=10, color='blue', stat='density')
plt.title('Histogram and KDE Plot')
plt.xlabel('SepalLengthCm')
plt.ylabel('Density')
plt.show()
```

```python
sns.histplot(df['SepalWidthCm'], bins=10, color='skyblue', stat='density')
plt.title('Histogram')
plt.xlabel('SepalWidthCm')
plt.ylabel('Density')
plt.show()
```

```python
sns.kdeplot(df['PetalLengthCm'], color='red', fill=True)
plt.title('KDE Plot')
plt.xlabel('PetalLengthCm')
plt.ylabel('Density')
plt.show()
```
